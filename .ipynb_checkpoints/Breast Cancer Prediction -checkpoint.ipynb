## Predicting Breast Cancer with Maching Learning 

### Imports 

```python
import pandas as pd 
import numpy as np 
import seaborn as sb 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

###  Get Data 

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
# drop the 'Unamed' Column 
data.drop('Unnamed: 32',axis=1,inplace=True)
data.head()
```

### Relabeling the Target Column with a LabelEncoder

```python
#Encoding categorical data values for the diagnosis column 
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data.diagnosis= labelencoder.fit_transform(data.diagnosis.values)
data.head()
```

```python
# group by diagnosis now 
data.diagnosis.value_counts()
# see the column values were encoded to:
## 1 means patient has an M(Malignat ):  Cancer 
## 0 means  patient hasn a B(Begnin) : No Cancer 
```

### Split Data into Train and Test Set 

```python
#copy the dataframe for further use 
df=data.copy()
#Features 
X=df.drop('diagnosis',axis=1)
#Labels
y=df.pop('diagnosis')
# split into train and test set 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

### Training the Model and Make Predictions   

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# define a Function to train all models at once 
def train_models(X_train,Y_train):
  
     #Using Logistic Regression 
    lr = LogisticRegression(random_state = 0)
    lr.fit(X_train, Y_train)
    #Using KNeighborsClassifier 
    knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    knn.fit(X_train, Y_train)
    #Using SVC
    svc= SVC(kernel = 'rbf', random_state = 0)
    svc.fit(X_train, Y_train)
    #Using GaussianNB 
    nb = GaussianNB()
    nb.fit(X_train, Y_train)
    #Using DecisionTreeClassifier 
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    tree.fit(X_train, Y_train)
    #Using RandomForestClassifier method of ensemble class to use Random Forest Classification algorithm
    rfc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    rfc.fit(X_train, Y_train)

    return lr, knn, svc, nb, tree, rfc
```

```python
# train our model now 
models = train_models(X_train,Y_train)
```

```python
type(models)
```

```python
#print model accuracy on the training data.
print('[0]Logistic Regression Training Accuracy:', models[0].score(X_train, Y_train))
print('[1]K Nearest Neighbor Training Accuracy:', models[1].score(X_train, Y_train))
print('[3]Support Vector Machine Training Accuracy:',models[2].score(X_train, Y_train))
print('[4]Gaussian Naive Bayes Training Accuracy:', models[3].score(X_train, Y_train))
print('[5]Decision Tree Classifier Training Accuracy:', models[4].score(X_train, Y_train))
print('[6]Random Forest Classifier Training Accuracy:', models[5].score(X_train, Y_train))
```

```python
# we could see that the RFC Classifier has the highest training accuracy 
# let's see for the test results 
```

```python
dict_models={'LogReg':models[0],'KNN':models[1],'SVC':models[2],'GNB':models[3],
             'DT':models[4],'RFC':models[5]}
```

```python
# show the Results 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for key,model in dict_models.items():
    print('Model :',key)
    #predictions 
    preds=model.predict(X_test)
    #Check precision, recall, f1-score
    print( classification_report(Y_test,preds) )
    #Another way to get the models accuracy on the test data
    print( accuracy_score(Y_test, preds))
    print('\n')
```

### Submit the Results 

```python
# from the results above ,we see that the Random Forest Classifier has the highest accuracy 
# So, we will take its predictions and sbmit it
rfc=dict_models['RFC']
rfc_preds=rfc.predict(X_test)
results=pd.Series(rfc_preds)
#submit our predictions 
submission = pd.DataFrame({'Id':results.index, 'Diagnosis': results})
submission.set_index('Id',inplace=True)
submission.to_csv('breast_cancer_submission.csv', index=False)
print(" Submission  successfully saved!")
```
